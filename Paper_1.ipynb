```python
"""To check whether entered integer is odd or even."""

num = int(input("Enter an integer: "))

if num % 2 == 0:
    print(num, "is even.")
else:
    print(num,"is odd.")
```

```python
"""To check whether the integer list of parameters are between 1 to 100. """

def in_between_1_100(*nums):
    for x in nums:
        if not 1 <= x <= 100:
            return False
    else:
        return True
    
#Demostration
print(in_between_1_100(1,2,3)) #True
print(in_between_1_100(101, -1, 1)) #False
```

```python
"""To diplay unique entered strings."""

String_list = []

while True:
    strg = input("Enter a string: ")
    if strg == "":
        break

    String_list.append(strg)

#to list unique strings

Unique_strings = set(String_list)

print("String List:", String_list)
print("Unique:", sorted(Unique_strings))
```
